```python
%matplotlib inline 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('USA_Housing.csv')
print(data.shape)
data.head()
```

```python
X = data['Area Population'].values
Y = data['Price'].values

mean_X = np.mean(X)
mean_Y = np.mean(Y)
print("mean of X =", mean_X)
print("Mean of Y =", mean_Y)

n = len(X)

num = 0
den = 0

for i in range(n):
    num += (X[i]- mean_X)*(Y[i]- mean_Y)
    den += (X[i]- mean_X)**2

m = num/den
c = mean_Y - (m * mean_X)

print("Value of Slope m = ", m)
print("Value of intrcept c = ", c)
```

```python
plt.scatter(X,Y, color = 'b', label = 'Scatter Plot')
```

```python
y = m * X + c

plt.scatter(X,Y, color = 'b', label = 'Scatter Plot')
plt.plot(X,y, color = 'r', label = 'Regression Line')
plt.xlabel('Area Population')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
upp = 0
low = 0
for i in range(n):
    upp += (y[i] - mean_Y)**2
    low += (Y[i] - mean_Y)**2
R_sq = 1 - (upp/low)
print(R_sq)
```
